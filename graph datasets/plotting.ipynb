```python
# importing the necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
```

```python
# in previewing the document the delimeter is semicolons.
# we specifiy the delimeter when reading our document
my_dataframe = pd.read_csv(r"C:\Users\Admin\Desktop\MODCOM2\datascience\datsciencepdfs\project\graph datasets\bank-additional-full.csv", sep=";")
my_dataframe
```

```python
# we check the datatypes and other information about our columns

my_dataframe.info()
```

```python
# we perform data cleaning

# we only need 12 columns so we remove the other columns

# we use the iloc function

my_dataframe = my_dataframe.iloc[:,0:12]
my_dataframe
```

```python
# we check for empties

my_dataframe.isnull().sum()
```

```python
#  we now perform visuals
# 20 different graphs to analyse the dataset

# count plot(for categorical variables) = 7
# histogram(quantitative variables) = 3
# scatter plot(quantitative variables) = 2
# heat map(qualitative columns) = 4
# violin plot() = 3
# Stripplot() = 1

```

```python
# age graph specifically a histogam

age_graph = my_dataframe['age'].plot(kind='hist', bins=30,)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title("AGE FREQUENCY")
age_graph
```

```python
# 2.duration graph

# my_dataframe['duration'].plot(kind='hist',  title='duration')
# plt.gca().spines[['top', 'right',]].set_visible(False)

# kde removes the smooth curve
duration_graph = sbn.distplot(my_dataframe['duration'] , kde = False , bins=20)
duration_graph
plt.title("DURATION FREQUENCY")
```

```python
campaign_graph = sbn.distplot(my_dataframe['campaign'] , kde = False , bins=20 )
campaign_graph
plt.title("CAMPAIGN FREQUENCY")
```

```python
# JOB 

# 7.count plot showing different job distribution

plt.figure(figsize = (20,10))
job_graph = sbn.countplot(x="job" , data = my_dataframe , palette="Blues")
job_graph

plt.title("JOB DISTRIBUTION")
plt.xlabel("JOB CATEGORIES")
plt.ylabel("JOB COUNT")
```

```python

# 8.count plot showing different modes of contact

plt.figure(figsize = (10,10))

contact_graph = sbn.countplot(x = "contact" , data = my_dataframe , palette = "Greens")
contact_graph



plt.title("CONTACT DISTRIBUTION")
plt.xlabel("TYPES OF CONTACT")
plt.ylabel("CONTACT COUNT")
```

```python
#9. age vs marital status

age_marital_graph = sbn.barplot(x="age" , y="marital" , data=my_dataframe , palette="Reds")
age_marital_graph

plt.title("AGE VS MARITAL STATUS")
plt.ylabel("MARITAL STATUS")
plt.xlabel("AGE")


# most people of ages 40 and above are mostly married or divorced
```

```python
# 10. age vs loan

# which age has taken the most loan

age_loan_graph = sbn.barplot(x="age" , y="loan" , data=my_dataframe , palette="plasma" , estimator=np.mean)
age_loan_graph

plt.title("AGE VS LOAN")
plt.xlabel("AGE")
plt.ylabel("LOAN STATUS")

# Graph suggests that there is equal distribution of loan between various ages
```

```python
# 12.Duration vs Campaign

duration_campaign_graph = sbn.scatterplot(x="campaign" , y="duration" , data=my_dataframe , palette="Oranges")
duration_campaign_graph

plt.title("DURATION VS CAMPAIGN")
plt.xlabel("CAMPAIGN")
plt.ylabel("DURATION")


```

```python
# cmaps(color maps)
# we have differernt color maps
# the default colormap for seaborn and matplotlib is viridis
# other colormaps for heatmaps (sequential) ie low to high are =>
#  YlGn
#  YlGnBu
#  YlOrRd
#  viridis
#  cividis

# colormaps(quantitative) are =>

# Pastel1
# Pastel2
# paired
# Set1
# Set2
# Set3
# Accent

# colormaps(diverging)ie moves away from a central area =>

# PiYG
# PRGn
# BrBG
# PuOr
# RdGy
# RdBu
# RdYlBu

```

```python
#13. heat map showing marital status vs education

cross_tab = pd.crosstab(my_dataframe['marital'] , my_dataframe['education'])


figure = plt.figure(figsize=(10,5))
educaion_marital_graph = sbn.heatmap(data=cross_tab , cbar=True , cmap="Paired")
educaion_marital_graph

plt.title("HEATMAP OF EDUCATION VS MARITAL STATUS")
# Yl -yellow
# Or = orange
# Rd = red

# the purpose of creating a cross tab is so that we can summarize our qualitative columns.After summarizing it we create a hitmap
```

```python
# 14. a heatmap to show based on level of education which job do they do

# create a cross table since the columns are quantitative

cross_table = pd.crosstab(my_dataframe['job'] , my_dataframe['education'])
# cross_table


# figure size
figure = plt.figure(figsize=(10,7))
# a heatmap to now visually represent the above crosstable

education_job_graph = sbn.heatmap(data=cross_table , cmap="Set1" , cbar=True)
education_job_graph

plt.title("HEATMAP SHOWING LEVEL OF EDUCATION VS JOB CATEGORY")




# pairplot
```

```python
# 15. Heatmap showing housing status and loan status

cross_table = pd.crosstab(my_dataframe['loan'] , my_dataframe['housing'])

figure = plt.figure(figsize=(12,7))

housing_lan_heatmap = sbn.heatmap(data=cross_table , cmap="Accent" , cbar=True )
housing_lan_heatmap

plt.title("HEATMAP SHOWING HOUSING STATUS AND LOAN STATUS")
```

```python
#16. Heatmap showing the relationship between housing and default

housing_default_crosstab = pd.crosstab(my_dataframe['housing'] , my_dataframe['default'])


figure = plt.figure(figsize=(10,5))
housing_default_graph = sbn.heatmap(data=housing_default_crosstab , cbar=True , cmap="Paired")
housing_default_graph


plt.title("HEATMAP MAP SHOWING RELATIONSHIP BETWEEN HOUSING AND DEFAULT")
```

```python
#17 Violin plot showing relationship between Marital status and age

figure = plt.figure(figsize=(10,5))

marital_age_violinplot = sbn.violinplot(x="marital" , y="age" , data=my_dataframe , palette="viridis")
marital_age_violinplot

plt.title("VIOLIN PLOT SHOWING THE RELATIONSHIP BETWEEN MARITSL AND AGE")
```

```python
# 18. Violin plot showing the relationship between education and age



figure = plt.figure(figsize=(10,7))
education_age_violinplot = sbn.violinplot(x="age" , y="education" , data=my_dataframe , palette="cividis")
education_age_violinplot

plt.title("VIOLIN PLOT SHOWING RELATIONSHIP BETWEEN EDUCATION AND AGE")
```

```python
# 19. Violin plot showing relationship between housing and age


figure = plt.figure(figsize=(10,7))
housing_age_violinplot = sbn.violinplot(x="age" , y="housing" , data=my_dataframe , estimator = np.count(), palette="viridis")
housing_age_violinplot

plt.title("VIOLIN PLOT SHOWING RELATIONSHIP BETWEEN HOUSING AND AGE")
```

```python
# Swarm plot showing relationship between age and marital status with a hue fo loan


figure = plt.figure(figsize=(10,7))
age_marital_swarmplot = sbn.stripplot(x="marital" , y="age" , hue='loan' , data=my_dataframe, palette="viridis")
age_marital_swarmplot

plt.title("STRIP PLOT SHOWING RELATIONSHIP BETWEEN AGE AND MARITAL STATUS")
```
